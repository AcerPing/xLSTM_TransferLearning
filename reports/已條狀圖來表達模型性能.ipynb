```python
from os import path, walk, listdir
import json

import matplotlib.pyplot as plt
import numpy as np
```

<h3> 讀取結果檔案 </h3>

<h4> Pre-Train 預訓練結果檔 </h4>

```python
def read_prediction_errors(file_path):
    """
    讀取 params.json 文件中的預測誤差數據。
    Args:
        file_path (str): JSON 文件的路徑。
    Returns:
        dict: 預測誤差數據的字典。
    """
    with open(file_path, 'r') as file:
        data = json.load(file)
    
    # 提取預測誤差指標
    error_metrics = {
        "MSE Loss": data.get("MSE Loss"),
        "RMSE Loss": data.get("RMSE Loss"),
        "MAE Loss": data.get("MAE Loss"),
        "R2 Score": data.get("R2 Score")
    }
    return error_metrics
```

```python
df_ponds_errors = dict()

# source
for i in range(4, 1, -1):
    file_path = f'./IoTpond_Result/pre-train/FishAquaponics_IoTpond{i}/params.json'
    errors = read_prediction_errors(file_path)
    df_ponds_errors[f'IoTpond{i}'] = errors

df_ponds_errors.keys() # 有哪些資料集。
```

<h3> Target DataSet 目標檔案 </h3>

```python
file_path = f'./IoTpond_Result/without-transfer-learning/FishAquaponics_IoTpond{1}/params.json'
errors = read_prediction_errors(file_path)
df_ponds_errors[f'IoTpond{1}'] = errors
```

```python
df_ponds_errors
```

```python
# 指標名稱（根據 JSON 文件中的鍵順序）
metrics = ["MSE Loss", "RMSE Loss", "MAE Loss"]
metrics.reverse()
metrics.append("R2 Score")
metrics
```

```python
datasets = list(df_ponds_errors.keys())  # 數據集名稱
datasets
```

```python
# 使用 get 方法提取 values，並為缺失值設置默認值（例如 0 或 None）
values = [
    [dataset.get(metric, 0) for metric in metrics]  # 為每個數據集提取指定的指標數值
    for dataset in df_ponds_errors.values()
]
values
```

<h4> 數據可視化 </h4>

```python
# 顏色設定
colors = ["#9370DB", "#2ca02c", "#ff7f0e", "#1E90FF"]  # MSE, RMSE, MAE, R2 分別使用藍、綠、橙、紅色
```

```python
# 設置橫條圖參數
y = np.arange(len(datasets))  # y 軸座標
bar_height = 0.2


# 創建圖表
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 8), gridspec_kw={"width_ratios": [2, 1]})

# 誤差指標 (MSE, RMSE, MAE)
for i, metric in enumerate(metrics[:-1]):  # 排除 R2
    ax1.barh(y + i * bar_height, [v[i] for v in values], bar_height, color=colors[i], label=metric)
    for j, val in enumerate([v[i] for v in values]):  # 添加標籤
        ax1.text(val + 0.001, y[j] + i * bar_height, f"{val:.4f}", va="center", fontsize=9)

ax1.set_xlabel("Error Metrics", fontsize=12)
ax1.set_title("Error Metrics Across Datasets", fontsize=14)

# 設置 y 軸標籤並修改特定標籤顏色
yticks = y + bar_height
ax1.set_yticks(yticks)
ax1.set_yticklabels(datasets, fontsize=10)
for label, dataset in zip(ax1.get_yticklabels(), datasets):
    label.set_color("red" if dataset == "IoTpond1" else "black")  # 如果是 IoTpond1，設置為紅色


ax1.legend(title="Error Metrics", fontsize=12, loc="best", frameon=True, edgecolor='black', fancybox=True)
# ax1.set_xlim(0, 0.12)


# 模型性能指標 (R2 Score)
r2_values = [v[-1] for v in values]  # 提取 R2 值
ax2.barh(y, r2_values, bar_height, color=colors[-1], label="R2 Score")
for i, val in enumerate(r2_values):  # 添加標籤
    ax2.text(val + 0.02, y[i], f"{val:.4f}", va="center", fontsize=9, color=colors[-1])

ax2.set_xlabel("R2 Score", fontsize=12)
ax2.set_title("R2 Score Across Datasets", fontsize=14)
ax2.set_yticks(y) # + bar_height
ax2.set_yticklabels([])
ax2.legend(fontsize=12, loc="lower right", frameon=True, edgecolor='black', fancybox=True)
ax2.set_xlim(0, 1)

# 調整布局
plt.tight_layout()
plt.savefig("./IoTpond_Result/ErrorMetrics_and_R2Score_Comparison.png", bbox_inches='tight')  # 儲存圖片，解析度為300 DPI
plt.show()
```

※ R2 Score 明顯區分了不同數據集的性能，突出了重要指標。
