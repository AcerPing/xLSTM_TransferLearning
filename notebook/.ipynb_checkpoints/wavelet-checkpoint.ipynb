```python
import numpy as np
from scipy import signal # 提供信號處理工具，如小波變換和高斯脈衝等。
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-1, 1, 200, endpoint=False) # 生成一個從 -1 到 1 的等間距數列，共包含 200 個點。
sig  = np.cos(2 * np.pi * 7 * t) + signal.gausspulse(t - 0.4, fc=2) # 將正弦波和高斯脈衝相加，得到複雜的信號。
widths = np.arange(1, 31) # 設置小波的尺度（scale），範圍為 1 到 30。
```

```python
plt.plot(sig) # 繪製原始信號的時域波形。
```

```python
cwtmatr = signal.cwt(sig, signal.ricker, widths)
```

```python
cwtmatr.shape # 表示 30 個尺度和 200 個時間點。
```

```python
plt.imshow(cwtmatr, extent=[-1, 1, 1, 31], cmap='PRGn', aspect='auto',
           vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())
plt.show()
```
